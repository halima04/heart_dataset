```python
import pandas
import seaborn
import matplotlib.pyplot as plt
pandas.options.display.max_columns = None
pandas.options.display.max_rows = None

csv_data = pandas.read_csv("heart_2020_cleaned.csv")
csv_data.head(25)
```

As the dataset is too big, I am using a small size of the sample. Otherwise, I will get "nan" over and over again.

```python
# What is the number of rows and columns in our data?
csv_shape = csv_data.head(25).shape
csv_shape
```

```python
# How would the data be described?
csv_description = csv_data.head(25).describe()
csv_description
```

```python
# How many of the first 25 respondents have heart disease and smoke?
disease_smoke = csv_data.head(25) [ (csv_data.head(25)["Smoking"].notna()) & (csv_data.head(25)["HeartDisease"] =="Yes") ].shape[0]
disease_smoke
```

```python
# How many of the first 25 resondents get between 6 to 8 hours of sleep?
sleep_time = csv_data.head(25) [ (csv_data.head(25) ["SleepTime"] >=6) & (csv_data.head(25) ["SleepTime"] <= 8) ].shape[0]
sleep_time
```

```python
# How many of the first 25 respondents have asthma?
have_asthma = csv_data.head(25) [ csv_data.head(25)["Asthma"] =="Yes" ].shape[0]
have_asthma
```

```python
# Out of the 25 respondents, how did most of them say their general health status is?
gen_health = csv_data.head(25) 

seaborn.barplot(data = gen_health, x = gen_health.index, y = "GenHealth")
fig = plt.gcf()
fig.set_size_inches(9,6)
plt.title("General Health Responses")
plt.show()
```

```python
# What is the correlation between mental and physical health of the first 25 respondents?
mental = csv_data.head(25)["MentalHealth"]
physical = csv_data.head(25)["PhysicalHealth"]

seaborn.scatterplot(x = mental, y = physical)
plt.show()
```
